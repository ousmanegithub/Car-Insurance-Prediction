### Étape 3 : Analyse et stratégies d'imputation des valeurs manquantes  
Maintenant que nous avons identifié les variables affectées par des données manquantes (Job, Education, Communication, Outcome), nous devons choisir une stratégie d'imputation (remplacement) pour chacune d'elles. Le choix de la méthode doit être justifié par le type de donnée (numérique ou catégorielle) et la nature de l'information manquante.

---

#### 1. Variable **Job** (Profession du client)

- **Nombre de valeurs manquantes** : 19 / 4000  
- **Pourcentage** : moins de 0,5 %

**Hypothèse sur la raison des données manquantes**  
L’employé de banque qui a rempli le formulaire n’a peut-être pas jugé pertinent de noter la profession du client, ou le client n’a pas voulu la communiquer. Étant donné le très faible nombre de manquants (19 cas isolés), il s’agit très probablement d’oublis ponctuels lors de la saisie ou de refus individuels de réponse.

**Stratégie d’imputation potentielle (hypothèse métier)**  
On pourrait raisonnablement supposer que ces 19 clients ont un profil socio-professionnel similaire à la très grande majorité de la clientèle de la banque.

**Méthode alternative sérieusement envisagée**  
Remplacer simplement par la profession la plus courante dans l’échantillon, c’est-à-dire le **mode** dans ce dataset.

**Décision finale pour le projet**  
Remplacement par le **mode** (modalité la plus fréquente).

#### 2. Variable **Education** (Niveau d’éducation)

- **Nombre de valeurs manquantes** : 169 / 4000  
- **Pourcentage** : ~4,2 %

**Hypothèse sur la raison des données manquantes**  
Le niveau d’éducation est une donnée relativement sensible (certaines personnes n’aiment pas dévoiler leur parcours scolaire) ou n’a pas été collecté de façon systématique à l’époque de la constitution du fichier. L’absence peut aussi être liée à l’âge (clients très âgés dont le système scolaire était différent) ou à la situation d’emploi (retraités, chômeurs, etc.).

**Stratégie d’imputation potentielle (hypothèse métier)**  
On pourrait essayer de **prédire le niveau d’éducation manquant** en se basant sur deux variables fortement corrélées :  
- l’âge du client  
- la profession (Job)  
Ces deux variables sont généralement de très bons prédicteurs du niveau de diplôme.

**Méthode alternative sérieusement envisagée**  
Remplacer simplement par le niveau d’éducation le plus fréquent dans l’échantillon (le **mode**).

**Décision finale pour le projet**  
Remplacement par le **mode** (modalité la plus fréquente).

#### 3. Variable **Communication** (Canal de communication utilisé pour le contact actuel)

- **Nombre de valeurs manquantes** : 902 / 4000  
- **Pourcentage** : ~22,5 %

**Hypothèse sur la raison des données manquantes**  
La description officielle du dataset indique uniquement deux canaux possibles : « cellular » et « telephone ».  
Quand cette colonne est vide (NaN), cela signifie très probablement qu’**aucun de ces deux canaux n’a été utilisé** pour contacter le client (contact par email, courrier, passage en agence, ou tout simplement aucun contact téléphonique tenté).  
**L’absence de valeur n’est donc pas un oubli : c’est une information en soi** (le client n’a pas été joint par téléphone).

**Stratégie d’imputation potentielle (hypothèse métier)**  
Créer une **nouvelle catégorie explicite** `'None'` ou `'unknown'` pour indiquer « absence de communication téléphonique ». Cela permet au modèle (et surtout au réseau de neurones) d’apprendre que l’absence de canal téléphonique est un signal prédictif très fort.

**Méthode alternative envisagée (à éviter)**  
Remplacer par le mode (« cellular ») → **ce serait une grave erreur** car on masquerait complètement l’information extrêmement discriminante « pas de contact téléphonique ».

**Décision finale pour le projet**  
Malgré la forte recommandation métier de créer une catégorie `'None'`, nous suivons la consigne de **simplicité maximale** → remplacement par le **mode**.

#### 4. Variable **Outcome** (Résultat de la campagne précédente)

- **Nombre de valeurs manquantes** : 3042 / 4000  
- **Pourcentage** : plus de 75 %
Qui pourrait etre aussi supprimer vue le pourcentage de ces valeurs manquantes mais acceptable pour nous comme on definis un seuil de (80%)
**Hypothèse sur la raison des données manquantes**  
C’est la variable la plus manquante du dataset.  
Cette absence coïncide **à 100 %** avec les lignes où la variable **DaysPassed = -1**, qui signifie officiellement « aucun contact précédent ».  
Donc, quand il n’y a jamais eu de campagne antérieure, il ne peut pas y avoir de résultat → **Outcome est forcément manquant**.  
Il ne s’agit pas d’un oubli : c’est une absence parfaitement déterministe et extrêmement porteuse de sens.

**Stratégie d’imputation potentielle (hypothèse métier)**  
Remplacer toutes les valeurs manquantes par une catégorie dédiée telle que **`'No_prev_contact'`** ou **`'none'`**. Cela permet au modèle de distinguer clairement les « prospects froids » (jamais contactés) des autres.

**Méthode alternative envisagée (à éviter)**  
Remplacer par le mode (« unknown » ou « failure ») → introduirait un bruit massif et masquerait la différence fondamentale entre clients déjà démarchés et clients totalement nouveaux.

**Décision finale pour le projet**  
Malgré l’évidence métier écrasante, nous respectons la consigne de **simplicité et standardisation** → remplacement par le **mode**  dans ce dataset).

### Décision globale finale 

> Bien que nous ayons émis des hypothèses très solides et détaillées sur la signification réelle de ces données manquantes (surtout pour **Communication** et **Outcome**), nous ne disposons pas d’une confirmation officielle de la banque sur la méthode exacte de collecte.  
> Pour garantir un processus **simple, reproductible et parfaitement standardisé**, pour des besoins d'automatisation dans le future,  
> **nous remplaçons toutes les valeurs manquantes des variables catégorielles par la modalité la plus fréquente (le mode)** à l’aide de `SimpleImputer(strategy='most_frequent')`.

### Tableau récapitulatif final des stratégies d'imputation retenues

| Variable         | Nb valeurs manquantes | % manquant | Hypothèse principale sur l’absence                                                                                  | Meilleure stratégie métier (recommandée en pratique)                 | Stratégie réellement appliquée | |                                     |
|------------------|-----------------------|------------|----------------------------------------------------------------------------------------------------------------------|-----------------------------------------------------------------------|--------------------------------------------------|-----------------------|--------------------------------------------------|
| **Job**          | 19                    | 0,48 %     | Oubli de saisie ou refus du client                                                                                  | Remplacement par le **mode**                                          | Remplacement par le **mode**                      `SimpleImputer(strategy='most_frequent')`        |
| **Education**    | 169                   | 4,23 %     | Donnée sensible ou non collectée systématiquement                                                                   | Création catégorie `'unknown'` (ou prédiction via Job + Age)          | Remplacement par le **mode**                      `SimpleImputer(strategy='most_frequent')`        |
| **Communication**| 902                   | 22,55 %    | **Absence de contact téléphonique** (canal non utilisé)                                                             | Création catégorie **`'none'`**                                       | Remplacement par le **mode**                      `SimpleImputer(strategy='most_frequent')`        |
| **Outcome**      | 3042                  | 76,05 %    | **Aucun contact précédent** (DaysPassed = -1 → absence déterministe)                                                | Création catégorie **`'no_previous_contact'`**                        | Remplacement par le **mode**                      `SimpleImputer(strategy='most_frequent')`        |

```python
imputer = SimpleImputer(strategy='most_frequent')
columns_cat_na = ['Job', 'Education', 'Communication', 'Outcome']
df[columns_cat_na] = imputer.fit_transform(df[columns_cat_na])
print("Imputation terminée : toutes les valeurs manquantes remplacées par le mode.")
```

### Étape 4 : Ingénierie de Variables (feature ingeneering) - Transformer le temps en un indicateur pertinent

Cette étape est fondamentale dans notre processus. Il s'agit de **l'ingénierie de variables**, où nous créons de nouvelles données plus intelligentes à partir des données brutes fournies, améliorant ainsi la capacité de notre futur modèle à prédire la cible `CarInsurance`.

Nous avons effectué les transformations suivantes :

#### 1. Comprendre et calculer la durée de l'appel (`CallDuration`)

Les colonnes `CallStart` et `CallEnd` contenaient simplement du texte (ex: "10:30:00"). Pour un ordinateur, ces chiffres et ces symboles n'ont aucune signification temporelle.

*   **Action :** Nous avons converti ces textes en un format "date/heure" que Python comprend, puis nous avons calculé la différence entre l'heure de fin et l'heure de début.
*   **Résultat :** Une nouvelle colonne `CallDuration` exprimée en minutes.

#### 2. L'importance prédictive de la durée

La durée de l'appel est un **indice comportemental clé**.

*   Un appel court (ex: 30 secondes) indique souvent un désintérêt ou un raccrochage.
*   Un appel long (ex: 10 minutes) indique que le client a posé des questions, écouté les offres et montré de l'intérêt.

En transformant les heures de début/fin peu utiles en une `CallDuration` numérique, nous donnons au réseau de neurones une information directe et puissante pour sa prédiction.

#### 3. Nettoyage final du jeu de données

Nous avons ensuite "fait le ménage" dans notre tableau de données :

*   **Suppression des heures de début/fin :** Elles sont inutiles maintenant que nous avons la durée.
*   **Suppression de l'`Id` :** Le numéro d'identification d'un client est arbitraire et n'aide en rien à prédire son comportement d'achat.

Notre jeu de données est maintenant optimisé, prêt et propre pour passer à l'étape suivante : la matrice de corrélation et, enfin, la construction du modèle.


```python
df['CallStart_dt'] = pd.to_datetime(df['CallStart'], format='%H:%M:%S')
df['CallEnd_dt'] = pd.to_datetime(df['CallEnd'], format='%H:%M:%S')

call_duration_timedelta = df['CallEnd_dt'] - df['CallStart_dt']

df['CallDuration'] = call_duration_timedelta.dt.total_seconds() / 60

print("Aperçu du DataFrame avec la nouvelle colonne CallDuration :")
df[['CallStart', 'CallEnd', 'CallDuration']].head()
```

```python
columns_to_drop = ['Id', 'CallStart', 'CallEnd', 'CallStart_dt', 'CallEnd_dt']

df.drop(columns=columns_to_drop, inplace=True)

print(f"\nColonnes restantes dans le DataFrame : {df.columns.tolist()}")
print(f"Nombre total de variables après nettoyage : {len(df.columns)}")
```

## Étape 6 : Analyse de la Corrélation entre Variables Numériques

### Objectif de l'analyse
Après avoir examiné l'influence individuelle de chaque variable sur la souscription, nous devons maintenant analyser les **relations entre les variables numériques elles-mêmes** pour détecter d'éventuels problèmes de **multicolinéarité**.

Pour que notre modèle d'Intelligence Artificielle fonctionne bien, il faut s'assurer que chaque donnée lui apporte une information unique. Si on lui donne deux fois la même information sous une forme légèrement différente, le modèle peut être confus.


### Pourquoi cette étape est important ?
Imaginez qu'on donne au modèle la "durée d'appel en minutes" ET la "durée d'appel en secondes". Ces deux chiffres disent exactement la même chose ! C'est ce qu'on appelle la redondance.

### Conséquence
La multicolinéarité se produit lorsque plusieurs variables d'entrée sont fortement corrélées entre elles. Dans ce cas :
- Le modèle peut devenir instable et difficile à interpréter
- Les performances de prédiction peuvent être affectées
- Le réseau de neurones peut attribuer de l'importance de manière incorrecte aux caractéristiques


### Méthodologie employée
Nous utiliserons une **matrice de corrélation de Pearson** visualisée sous forme de **heatmap (carte de chaleur)**. Cette représentation graphique permet d'identifier rapidement les relations linéaires entre variables.

### Échelle d'interprétation des coefficients de corrélation
- **|r| ≥ 0.7** : Corrélation forte (risque de redondance élevé)
- **0.3 ≤ |r| < 0.7** : Corrélation modérée (à surveiller)
- **|r| < 0.3** : Corrélation faible (variables relativement indépendantes)
- **r ≈ 0** : Pas de corrélation linéaire détectable

## Résultats attendus de cette analyse
1. **Identification des paires de variables redondantes** pouvant être simplifiées
2. **Validation des observations qualitatives** faites lors des visualisations précédentes
3. **Sélection des variables numériques les plus informatives** pour notre modèle
4. **Confirmation des corrélations avec la variable cible** (*CarInsurance*)

```python
df_number = df.select_dtypes('number')
binary_cols = ['CarLoan', 'Default', 'HHInsurance', 'Id']
numerical_cols_to_plot = [col for col in df_number.columns if col not in binary_cols + ['CarInsurance']]

df_correlation = df[numerical_cols_to_plot]

corr_matrix = df_correlation.corr()

plt.figure(figsize=(12, 8))

sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Matrice de Corrélation des Variables Numériques Pures')
plt.show()
```

La carte de chaleur (heatmap) ci-dessus nous permet de voir d'un coup d'œil si nos variables numériques sont liées entre elles. Nous cherchons des indices de redondance (doublons d'information).

### Comment lire la carte ?
- **Les cases rouges** indiquent une très forte liaison (valeur proche de 1 ou -1).
- **Les cases bleues** indiquent peu ou pas de liaison (valeur proche de 0).

### Ce que nous observons :

Cette valeur est logique et n'est pas assez élevée (pas proche de 1) pour créer un problème de redondance majeur pour notre modèle. Nous pouvons garder les deux variables.

### 2. LastContactDay et NoOfContacts (Corrélation de 0.14)
Une très faible liaison est présente.

### 3. Toutes les autres paires (proches de 0.0)
Toutes les autres variables (Age, Balance, etc.) sont presque indépendantes les unes des autres.

### Conclusion de l'analyse :
Notre analyse de corrélation confirme qu'il n'y a pas de redondance significative entre nos variables numériques. C'est une très bonne nouvelle, car cela signifie que chaque variable apporte une information unique au modèle d'intelligence artificielle.

## Étape 7 : Encodage des variables qualitative

Jusqu'à présent, certaines de nos données sont sous forme de texte (variables catégorielles), par exemple :
- **Job** : 'management', 'technician', 'blue-collar'
- **Marital** : 'married', 'single', 'divorced'
- **Communication** : 'cellular', 'telephone', 'unknown'

**Le problème :**
Un réseau de neurones (et la plupart des algorithmes d'Intelligence Artificielle) est un outil mathématique. Il ne sait pas "lire" le texte. Si nous lui donnons des mots, il ne comprendra pas comment les utiliser dans ses calculs pour faire une prédiction.

### La solution : L'encodage One-Hot (OneHotEncoder)

L'encodage One-Hot est une technique qui permet de convertir ces informations textuelles en un format purement numérique (des chiffres 0 ou 1) que le modèle peut comprendre, sans introduire de notion d'ordre ou de hiérarchie.

### Comment ça marche ?

Pour chaque colonne de texte, nous créons de nouvelles colonnes, une pour chaque option possible. Par exemple, pour la variable **Marital** qui a trois options :

| Avant Encodage | Marital_Married | Marital_Single | Marital_Divorced |
|----------------|-----------------|----------------|------------------|
| married        | 1               | 0              | 0                |
| single         | 0               | 1              | 0                |
| divorced       | 0               | 0              | 1                |

Chaque client a maintenant un **"1"** dans la colonne qui lui correspond et des **"0"** partout ailleurs. C'est comme **cocher une case**.

### L'importance cruciale (drop='first')

Dans le code ci-dessous, nous utilisons une astuce importante : `drop='first'`. Nous supprimons la première colonne créée (par exemple, `Marital_Married`).

**Pourquoi ?** Si un client a **0** dans `Marital_Single` **ET 0** dans `Marital_Divorced`, le modèle sait automatiquement qu'il est **"Married"**. Avoir la troisième colonne serait redondant. Cela aide le modèle à être plus efficace et à éviter certains problèmes mathématiques.

Nous utilisons la classe `ColumnTransformer` de sklearn pour appliquer cette logique uniquement aux colonnes textuelles, tout en laissant les colonnes numériques intactes.

```python
X_pre_encode = df.drop('CarInsurance', axis=1)
y = df['CarInsurance'].values 

categorical_cols_X = X_pre_encode.select_dtypes(include=['object']).columns.tolist()
print(f"Colonnes catégorielles à encoder : {categorical_cols_X}")

ct = ColumnTransformer(
    [('one_hot_encoder', OneHotEncoder(drop='first', sparse_output=False), categorical_cols_X)],
    remainder='passthrough' 
)

X_encoded = ct.fit_transform(X_pre_encode)

print("\nEncodage One-Hot terminé.")
print(f"Forme de X (features) après OneHotEncoder : {X_encoded.shape}")
print(f"Forme de y (cible) : {y.shape}")
```

## Étape 8 : Séparation des Données d'Entraînement et de Test, et Normalisation (Scaling)

Cette étape est cruciale pour garantir que le modèle que nous construisons soit fiable et efficace lorsqu'il sera déployé pour aider les conseillers de la banque à cibler de nouveaux clients. Elle aborde deux concepts essentiels.

---

### Concept 1 : La Séparation des Données (Train/Test Split)
**Outil utilisé :** `train_test_split`

### Analogie de la campagne marketing :
Imaginez que le directeur marketing veuille évaluer l'efficacité de sa nouvelle stratégie d'assurance automobile.  
Il ne peut pas utiliser **tous les 4000 clients historiques** pour apprendre au modèle, puis prétendre que le modèle a réussi s'il "prédit" parfaitement ce qu'il vient de voir.  
Il doit être honnête sur la performance future.

**La solution est de diviser la liste en deux parties :**

1. **L'ensemble d'Entraînement (Train Set - 80% des clients)** :  
   C'est la base de données sur laquelle le modèle va apprendre les schémas (par exemple, "les appels longs au mois de mars fonctionnent bien").

2. **L'ensemble de Test (Test Set - 20% des clients)** :  
   C'est l'échantillon **"mystère"**. Le modèle **n'y a pas accès** pendant son apprentissage.

### Importance pour la prédiction :
Cette séparation est vitale pour éviter le **surapprentissage (overfitting)**.  
Le surapprentissage se produit lorsque le modèle **mémorise** les données par cœur au lieu d'apprendre à **généraliser**.

L'ensemble de test simule l'arrivée de **nouveaux clients**. Les métriques calculées sur l'ensemble de test (comme le F1-score ou l'Accuracy) nous donnent une estimation **réaliste** de la performance du modèle dans le monde réel. C'est la seule façon de savoir si le modèle est vraiment utile pour prédire l'avenir.

---

### Concept 2 : La Normalisation ou Mise à l'échelle (Scaling)
**Outil utilisé :** `StandardScaler`

### Analogie du système de notation interne de la banque :
Imaginez que le réseau de neurones est un comité interne qui évalue la probabilité qu'un client souscrive à l'assurance. Ce comité se base sur plusieurs critères, notamment :
- **L'Âge** (valeurs typiques entre 18 et 80 ans).
- **Le Solde Bancaire** (valeurs typiques entre 0 € et 50 000 € ou plus).
- **La Durée d'Appel** (valeurs typiques entre 0 et 30 minutes).

### Le problème d'échelle :
Les chiffres du solde bancaire sont des centaines, voire des milliers de fois **plus grands** que ceux de l'âge ou de la durée d'appel.  
Si nous donnons ces chiffres bruts au modèle, lors de ses calculs initiaux, les grandes valeurs du solde bancaire vont **dominer** et masquer l'influence potentielle de l'âge ou de la durée d'appel, simplement parce que les chiffres sont plus grands.

### Importance pour la prédiction :
La normalisation résout ce déséquilibre.  
Le `StandardScaler` transforme toutes les variables pour qu'elles aient **la même échelle** d'importance. Chaque variable est **centrée autour de zéro**, avec une variation similaire.

Cela permet au réseau de neurones d'accorder une attention **égale** à l'âge, au solde, et à la durée d'appel au début de l'apprentissage. C'est ensuite au modèle de décider, par lui-même, quelles variables sont réellement les plus prédictives de la souscription.

**Cette étape est essentielle pour la stabilité et la rapidité de l'apprentissage du réseau de neurones.**

```python
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42, stratify=y)

print(f"Forme de l'ensemble d'entraînement (X_train): {X_train.shape}")
print(f"Forme de l'ensemble de test (X_test): {X_test.shape}")

sc = StandardScaler()
X_train_scaled = sc.fit_transform(X_train)
X_test_scaled = sc.transform(X_test)
```

# Étape 9 : Construction de l'Architecture du Réseau de Neurones

Dans cette étape, nous construisons la "machine intelligente" qui va apprendre à prédire l'achat d'assurance. Un réseau de neurones, c'est comme un **comité d'experts organisés en couches**.
- Chaque **"expert"** est un **neurone**.
- Chaque **"couche"** est un **niveau de réflexion**.

Nous utilisons la bibliothèque **Keras (dans TensorFlow)** pour assembler ce comité.

---

## Les Composants Clés de Notre Architecture

Notre modèle est construit en empilant des couches (`Sequential`) :

### 1. La Couche d'Entrée (Input Layer)
**Rôle :** C'est par ici qu'arrivent les données clients propres et normalisées (âge, durée d'appel, type de communication, etc.).  
**Dans le code :** Nous définissons `input_dim` qui correspond au **nombre total de caractéristiques** (colonnes) que nous avons pour chaque client.

### 2. Les Couches Cachées (Hidden Layers) - Les Niveaux de Réflexion
**Rôle :** Ce sont les couches intermédiaires où l'intelligence opère. Chaque neurone combine les informations reçues de la couche précédente, effectue un calcul, et transmet le résultat.

- **`units=10`** : Nous avons choisi 10 neurones pour chacune de nos deux couches cachées. C'est un nombre équilibré qui permet au réseau d'analyser suffisamment de motifs complexes (comme le fait qu'un appel long et un statut "célibataire" augmentent la probabilité d'achat) sans être trop gourmand en calcul.
- **`activation='relu'`** : C'est la **"fonction de réflexion"** la plus populaire. Elle décide si un neurone doit être **"actif"** ou **"inactif"** en fonction des informations reçues.
- **`Dropout(0.1)`** : C'est une sécurité ! Pendant l'apprentissage, nous **"éteignons" aléatoirement 10% des neurones**. Cela force le modèle à être plus robuste et à ne pas se reposer sur un seul chemin de décision, évitant ainsi le **surapprentissage**.

### 3. La Couche de Sortie (Output Layer) - La Décision Finale
**Rôle :** C'est la couche qui nous donne la réponse finale.

- **`units=1`** : Nous n'avons besoin que d'un seul neurone car la réponse est **binaire** : 1 (le client achète l'assurance) ou 0 (il n'achète pas).
- **`activation='sigmoid'`** : Cette fonction est parfaite pour la classification. Elle **"écrase"** le résultat final en une **probabilité unique**, comprise entre 0 et 1 (par exemple, 0.85 signifie 85% de chances de souscrire).

---

## La Compilation : Les Règles du Jeu

Une fois l'architecture définie, nous devons dire au modèle **comment apprendre** :

- **`loss='binary_crossentropy'`** : C'est la **"fonction de coût"** ou la mesure d'erreur. Elle dit au modèle à quel point il s'est trompé. L'objectif de l'apprentissage sera de **minimiser cette erreur**.
- **`optimizer='adam'`** : C'est l'algorithme qui dirige l'apprentissage. Il dit au modèle comment ajuster ses **"poids"** (la force des connexions entre neurones) pour réduire l'erreur de manière efficace.
- **`metrics=['accuracy']`** : C'est la métrique que nous allons suivre pendant l'entraînement pour voir la progression (le **taux de bonnes réponses**).

```python
Classifier = Sequential()

input_dim_value = X_train_scaled.shape[1] 

print(f"Notre couche d'entrée a {input_dim_value} variables.")

Classifier.add(Dense(units=19, activation='relu', kernel_initializer='uniform', input_dim=input_dim_value))
Classifier.add(Dropout(0.1)) 

Classifier.add(Dense(units=10, activation='relu', kernel_initializer='uniform'))
Classifier.add(Dropout(0.1))

Classifier.add(Dense(units=1, activation='sigmoid', kernel_initializer='uniform'))

Classifier.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

print("\nArchitecture du réseau de neurones construite et compilée.")
Classifier.summary()
```

# Étape 10 : Entraînement du Réseau de Neurones

C'est l'étape la plus attendue : nous allons donner nos données d'entraînement au réseau de neurones que nous venons de construire. C'est le moment où le **"comité d'experts"** commence à apprendre de ses erreurs.

---

### Comment se déroule l'apprentissage ?

L'apprentissage se fait par **essais et erreurs**, selon les règles que nous avons définies lors de la compilation (avec l'optimiseur `adam` et la fonction de coût `binary_crossentropy`).

**Le cycle d'apprentissage :**
1. Le modèle fait une **prédiction initiale** (souvent fausse au début).
2. Il calcule son **erreur (loss)**.
3. Puis, il ajuste ses connexions internes (ses **"poids"**) pour faire une **meilleure prédiction** la prochaine fois. C'est le principe de la **rétropropagation du gradient (backpropagation)**.

---

## Les Paramètres Clés (batch_size et epochs)

Nous contrôlons cet apprentissage avec deux paramètres principaux :

### 1. batch_size (Taille du lot = 10)

**Analogie de la formation bancaire :**
Imaginez former les conseillers de la banque à repérer les clients intéressés. Vous ne pouvez pas leur donner l'intégralité des 4000 dossiers d'un coup, ils seraient submergés.

**Vous leur donnez les dossiers par lots de 10 clients (`batch_size=10`).**
Après chaque lot de 10, vous leur faites un retour sur leurs erreurs.

**Importance :**
Apprendre par **petits lots** est plus efficace et stable que d'apprendre client par client, ou de tout apprendre d'un coup.

### 2. epochs (Nombre d'époques = 50)

**Analogie de la répétition :**
Une **"époque"** signifie que le modèle a vu l'intégralité de l'ensemble de données d'entraînement (nos 80% de clients) **une seule fois**.

**`epochs=50`** signifie que le modèle va refaire tout le processus **50 fois**.
À chaque époque, il devient un peu plus intelligent et réduit son erreur.

**Importance :**
Nous devons choisir un nombre **suffisant** d'époques pour que le modèle apprenne bien, mais **pas trop**, pour éviter le **surapprentissage** (mémorisation par cœur).

```python
print("\nDébut de l'entraînement du modèle...")

history = Classifier.fit(X_train_scaled, y_train,
                         batch_size=10,
                         epochs=50,
                         validation_split=0.1, 
                         verbose=1)

print("Entraînement terminé.")
```

# Étape 11 : Évaluation des Performances du Modèle

L'entraînement est terminé. Le modèle a étudié ses **80 % de données**. Il est maintenant temps de lui faire passer son **examen final** sur les **20 % de données qu'il n'a jamais vues** (`X_test_scaled`).

---

## Pourquoi évaluer sur l'ensemble de test ?

C'est l'étape de **vérité**. Le but est de répondre à la question : **"Le modèle sera-t-il utile lorsqu'il sera confronté à un nouveau client de la banque la semaine prochaine ?"**. L'évaluation sur l'ensemble de test nous donne une **estimation réaliste** de la performance future du modèle dans un scénario réel.

---

## Le Choix des Métriques : Au-delà de la simple Précision (Accuracy)

Nous utilisons **plusieurs indicateurs** pour évaluer le modèle. La simple **accuracy** (précision globale) peut être **trompeuse**.

### Exemple dans le contexte bancaire :
Si **95 % des clients disent "Non"** à l'assurance, un modèle qui dit **"Non" à tout le monde** aurait **95 % de précision** ! Pourtant, il serait **inutile** pour trouver les **5 % de clients qui disent "Oui"**.

C'est pourquoi nous utilisons des métriques plus fines :

1. **Précision (Precision)** :  
   *"Sur tous les clients que le modèle a prédit comme 'intéressés' (Oui), combien l'étaient réellement ?"*  
   → **Évite de faire perdre du temps aux conseillers avec de faux positifs.**

2. **Rappel (Recall) / Sensibilité** :  
   *"Sur tous les clients qui étaient réellement intéressés, combien le modèle a-t-il réussi à identifier ?"*  
   → **Évite de manquer des ventes potentielles.**

3. **F1-score** :  
   Un **score unique** qui combine la **Précision** et le **Rappel**. C'est souvent la **meilleure métrique** à regarder lorsque les données sont **déséquilibrées** (comme c'est le cas entre les "Oui" et les "Non" à l'assurance).

---

## Le Seuil de Décision (0.4)

Le modèle de sortie nous donne une **probabilité** (par exemple, 0.65). Par défaut, on considère que si c'est **supérieur à 0.5**, c'est "Oui".

**Nous ajustons ce seuil à 0.4.** Cela signifie que nous sommes prêts à prendre **un peu plus de risques** (plus de faux positifs) pour nous assurer de **ne manquer aucun client réellement intéressé** (augmenter le rappel).

```python
def eval_perf (model_to_evaluate, X_data, y_true):
    """
    Évalue le modèle en utilisant plusieurs métriques clés pour la classification binaire.
    """
    y_pred_probs = model_to_evaluate.predict(X_data)

    threshold = 0.4
    y_pred = (y_pred_probs > threshold).astype(int).flatten()
    
    print(f"\n--- Rapport de Classification (Seuil > {threshold}) ---")
    print(classification_report(y_true, y_pred))

    print("\nMatrice de Confusion :")

    print(confusion_matrix(y_true, y_pred))

eval_perf(Classifier, X_test_scaled, y_test)
```

# Étape 12 : Optimisation des Hyperparamètres

Jusqu'ici, nous avons fait des choix "raisonnables" pour les réglages de notre modèle (par exemple, 50 époques, taille de lot de 10, optimiseur 'adam'). Ces réglages sont appelés **hyperparamètres**.

---

## C'est quoi un hyperparamètre ?

Un hyperparamètre est un **paramètre de configuration** qui **n'est pas appris** par le modèle lui-même pendant l'entraînement, mais que **nous, humains, devons définir** avant de lancer l'entraînement.

## L'importance de cette étape :

Le choix de ces hyperparamètres a un **impact majeur** sur la performance finale de notre modèle. Un mauvais choix peut rendre un modèle **inutile**, tandis qu'un bon choix peut transformer un modèle **moyen** en un modèle **excellent**. L'optimisation vise à trouver **la "recette" parfaite** pour notre problème spécifique de prédiction d'assurance.

---

## L'Outil Utilisé : GridSearchCV (Recherche sur Grille)

### Comment ça fonctionne ?

**GridSearchCV** (Recherche sur Grille de Validation Croisée) est une méthode **systématique et exhaustive** pour trouver la meilleure combinaison d'hyperparamètres.

1. **Nous définissons une "grille" :** Nous fournissons une liste de valeurs que nous souhaitons tester pour chaque hyperparamètre.
   *Exemple : Tester `batch_size` [10, 15, 20] et `epochs` [30, 50, 80].*

2. **L'outil teste tout :** GridSearchCV va tester **toutes les combinaisons possibles** (ici, 3 * 3 = 9 modèles différents).

3. **Il évalue le meilleur :** Il entraîne chacun de ces 9 modèles et sélectionne celui qui donne le **meilleur F1-score** (la métrique que nous avons choisie pour notre projet).

---

## Comparaison avec RandomizedSearchCV (Recherche Aléatoire)

Il existe une autre méthode populaire, **RandomizedSearchCV**, qui choisit des combinaisons d'hyperparamètres de manière **aléatoire** au lieu de manière exhaustive.

| Caractéristique       | GridSearchCV                        | RandomizedSearchCV                  |
|-----------------------|-------------------------------------|-------------------------------------|
| **Méthode**           | Teste TOUTES les combinaisons       | Teste un échantillon aléatoire      |
| **Temps d'exécution** | Long (très)                         | Plus rapide                         |
| **Garantie du meilleur** | Trouve toujours le meilleur réglage de la grille | Peut manquer le meilleur réglage |
| **Utilisation**       | Grilles de taille raisonnable       | Grilles très grandes ou complexes   |

### Avantages de GridSearchCV pour notre projet :

1. **Exhaustivité garantie :** Puisque notre grille d'hyperparamètres reste relativement petite, GridSearchCV nous garantit de trouver **la combinaison optimale** parmi nos choix.
2. **Reproductibilité :** C'est une approche **systématique** facile à expliquer et à reproduire.

```python
def create_model(optimizer='adam'):
    model = Sequential()
    model.add(Dense(units=10, activation='relu', kernel_initializer='uniform', input_dim=input_dim_value))
    model.add(Dropout(0.1))
    model.add(Dense(units=10, activation='relu', kernel_initializer='uniform'))
    model.add(Dropout(0.1))
    model.add(Dense(units=1, activation='sigmoid', kernel_initializer='uniform'))
    model.compile(optimizer=optimizer, loss='binary_crossentropy', metrics=['accuracy'])
    return model

model = KerasClassifier(model=create_model, verbose=0)

param_grid = {
    "batch_size" : [5, 10, 20, 15, 30, 35, 40],  
    "epochs" :[30, 50, 80],      
    "optimizer" : ["adam", "rmsprop"] 
}

grid_search = GridSearchCV(estimator=model, param_grid=param_grid, scoring="f1", cv=3, verbose=1)

print("\nLancement de la recherche d'hyperparamètres (Grid Search)...")
grid_search = grid_search.fit(X_train_scaled, y_train)

print("\nMeilleurs hyperparamètres trouvés :")
print(grid_search.best_params_)

best_model_from_grid = grid_search.best_estimator_

print("\n--- Évaluation finale sur l'ensemble de TEST avec le modèle optimisé ---")
eval_perf(best_model_from_grid, X_test_scaled, y_test)
```
